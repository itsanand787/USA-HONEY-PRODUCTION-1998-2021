```python
plt.figure(figsize=(10,6))
x = q7['year']
y = q7['yieldpercol']
slope, intercept = np.polyfit(x, y, 1)
regression_line = slope * x + intercept
plt.scatter(x, y)
plt.plot(x, regression_line, c='r')
plt.xlabel('Yield_per_colony')
plt.ylabel('year')
plt.grid()
plt.title('Trend of honey yield per colony')
plt.show()
```

```python

honeyfinaldf = pd.read_csv('C:/Users/krana/Desktop/Python notes/honeyproduction 1998-2021.csv')
hny = honeyfinaldf.groupby('year').agg({'totalprod': 'sum', 'prodvalue': 'sum'}).reset_index()
hny.columns = ['year', 'totalproduction', 'VoP']

# Create subplots
fig, axs = plt.subplots(ncols=2, figsize=(12, 5))

# Plot line of best fit and raw data for total production
sns.regplot(data=hny, x='year', y='totalproduction', ax=axs[0], color='red', scatter_kws={'s':10})
sns.scatterplot(data=hny, x='year', y='totalproduction', ax=axs[0], color='blue', s=20)

# Plot line of best fit and raw data for value of production
sns.regplot(data=hny, x='year', y='VoP', ax=axs[1], color='green', scatter_kws={'s':10})
sns.scatterplot(data=hny, x='year', y='VoP', ax=axs[1], color='darkblue', s=20)

# Format plots
for ax, label in zip(axs, ['Total Honey Production', 'Value of Production']):
    ax.set_title(label, fontsize=14)
    ax.set_xlabel('Year', fontsize=12)
    ax.set_ylabel('Millions of Pounds', fontsize=12)
    ax.set_xticks(range(1998, 2013, 2))
    ax.tick_params(axis='both', which='major', labelsize=10)
    ax.grid(which='major', axis='both', linestyle='-', color='lightgrey')

# Set main title and adjust subplots
plt.suptitle('Total Honey Production Vs. Value of Production', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show plot
plt.show()
```

```python
# Create a line plot of total honey production
sns.lineplot(x='year', y='totalproduction', data=hny, color='red', linewidth=2)

# Create a line plot of value of production
sns.lineplot(x='year', y='VoP', data=hny, color='green', linewidth=2)

# Add scatter plot points for both lines
sns.scatterplot(x='year', y='totalproduction', data=hny, color='blue', s=20)
sns.scatterplot(x='year', y='VoP', data=hny, color='darkblue', s=20)

# Set the x-axis tick marks
plt.xticks(ticks=range())

# Set the plot title and labels
plt.title('Total Honey Production Vs. Value of Production')
plt.xlabel('Year')
plt.ylabel('')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the default font size
plt.rcParams.update({'font.size': 14})

# Set the figure size
plt.figure(figsize=(15, 10))

# Create a line plot of total honey production
sns.lineplot(x='year', y='totalproduction', data=hny, color='red', linewidth=2)

# Create a line plot of value of production
sns.lineplot(x='year', y='VoP', data=hny, color='green', linewidth=2)

# Add scatter plot points for both lines
sns.scatterplot(x='year', y='totalproduction', data=hny, color='blue', s=20)
sns.scatterplot(x='year', y='VoP', data=hny, color='darkblue', s=20)

# Set the x-axis tick marks
plt.xticks(ticks=hny['year'], labels=hny['year'])

# Set the plot title and labels
plt.title('Total Honey Production Vs. Value of Production')
plt.xlabel('Year')
plt.ylabel('Millions of Pounds')

# Add a legend to the plot
plt.legend(['Total Production', 'Value of Production'])

# Add a grid to the plot
plt.grid()

# Show the plot
plt.show()
```
